```python
#here, we have used parse_dates=['release_date'] to convert the datatype of ['release_date'] to datetime.
df = pd.read_csv("/content/drive/MyDrive/Colab Notebooks/imdb_data - imdb_data.csv", parse_dates=['release_date'])
df.info()
```

```python
#this will give us a series of boolean datatype, where wherever we have hindi,i.e, hi, there we'll get true for that column.
df['original_language'] == "hi"
```

```python
#this will give us the DataFrame of the above series.
df[df['original_language'] == "hi"]
```

```python
#this will filter the above DataFrame on the basis of  Columns ['budget','revenue','title'].
df[df['original_language'] == "hi"][['budget','revenue','title']]
```

```python
# Get the Movies were runtime is greater than 180
df.loc[df['runtime']>180].head()
```

```python
# Get the Movies were runtime is greater than 180 and original language is Hindi(hi)
df.loc[(df['runtime'] > 180) & (df['original_language'] == "hi"),['title','budget','revenue']]
```

```python
# Get the Movies were runtime between 150 to 180

df[(df['runtime'] >= 150) & (df['runtime'] <= 180)]
```

```python
# The code first adds a new column named 'Interval' to the DataFrame where intereval time is Half of runtime
df['Intereval'] = df['runtime']*0.5
df.head()
```

```python
#a new column named 'profit' is added to the DataFrame.
df['profit'] = df['revenue'] - df['budget']
df
```

```python
#.drop(): This is a method provided to remove rows or columns from the DataFrame.
#axis=1 indicates that you want to drop a column
df.drop(['Intereval'], axis = 1)
```

```python
#When inplace is set to True, as in this case, the original DataFrame will be modified, and no new DataFrame will be returned.
df.drop(['Intereval'],axis=1, inplace = True)
df
```

```python
#axis=0 indicates that rows are being targeted for removal.
df.drop([2999],axis=0)
```

```python
# Calculate the Sum,Mean and Mode of Revenue
df['revenue'].sum()
df['revenue'].mean()
df['revenue'].mode()
```

```python
# Calculate the Max Profit
max_profit=df['profit'].max()
max_profit
```

```python
# Get The Max Profited Movie Name
df.loc[df['profit']== max_profit,'title']
```

```python
# Display the Max Profited Movie with Revenue
df.loc[df['revenue']== df['revenue'].max(),['title','revenue']]
```

```python
# Display the Max budget and Lowest revenue with title
df.loc[(df['budget'] == df['budget'].max()) | (df['revenue'] == df['revenue'].min()),['title','budget','revenue']]
```

```python
# Display the name of movie who  have the max revenue
df[df['revenue'] == df['revenue'].max()]['title'][1126]
```

```python
# Give the Name of Movie how have Lowest budget
df.loc[df['budget']== df['budget'].min(),['title','budget']]
```

```python
# Set the imdb_id as a Index
df.set_index('imdb_id', inplace=True)
df.head()
```

```python
# Reset the Index
df.reset_index(drop=True)
```

```python
# Create the New Column of title with Uppercase
def to_uppercase(string):
  return string.upper()

df['new_title'] = df['title'].apply(to_uppercase) #note- the result can be seen in the last column
df
```

```python
# Sort the Value by runtime as Descending
df.sort_values('runtime',ascending=False,inplace=True)
df.head()
```

```python
# Sort the DataFrame by Index
df.sort_index(inplace= True)
```
